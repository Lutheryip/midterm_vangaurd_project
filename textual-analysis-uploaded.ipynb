```python
file_obj = open("C:/Users/LutherYip/Downloads/notes/topic 4/Fifty-Shades-Darker.txt")
file_contents = file_obj.readlines()
type(file_contents)
import string
punctuation = string.punctuation
import matplotlib.pyplot as plt
```

```python
print(file_contents[-2:-1])    
```

```python
book_contents = file_contents[23:-1]
```

```python
stop_words = open("C:/Users/LutherYip/Downloads/notes/topic 4/full-list-of-bad-words_text-file_2022_05_05.txt")
stop_contents = stop_words.readlines()
bad_word = []
for line in stop_contents:
    line_c = line.translate(str.maketrans("","",punctuation))
    line_cs = line.replace("\n","")
    bad_word.append(line_cs)
censored_words = bad_word[14:-1]
```

```python

book_cleaned_lines = []
for line in book_contents:
    line_c = line.translate(str.maketrans("","",punctuation))
    line_cl = line_c.lower()
    line_cls = line_cl.strip()
    book_cleaned_lines.append(line_cls)
print(book_cleaned_lines[0:30])
```

```python
word_counts = {}
for line in book_cleaned_lines:
    words = line.split()
    for word in words:
        if word in censored_words:
            if word in word_counts.keys():
                word_counts[word] += 1
            else:
                word_counts[word] = 1
#print(word_counts)
            
present_words = list(word_counts.keys())
type(present_words)
```

```python
word_counts_sorted = sorted(word_counts.items(),key=lambda x: x[1],reverse=True)
print(word_counts_sorted)
```

```python
word, count = zip(*word_counts_sorted)
import numpy as np
X = present_words[0:26]
top25_c = count[0:26]
#top25_d = count
X_axis = np.arange(26)

plt.figure(figsize=(20,30))
plt.barh(X_axis -0.2 ,top25_c,0.4,label ="series 1")
#plt.barh(X_axis + 0.2,top25_d, 0.4,label = "series 2")
plt.xlabel('count')
plt.ylabel('word')
plt.xticks(fontsize=10)
plt.yticks(X_axis, X,fontsize=10)
plt.show()
```
